```python
import pandas as pd
data = pd.read_csv('总表.csv')
data.info()
```

```python
pd.set_option('display.max_columns', None)
data.head()
```

```python
#去掉前缀
data['驱动电机控制器温度'] = data['驱动电机控制器温度'].str.split(':').str[1]
data['驱动电机转速'] = data['驱动电机转速'].str.split(':').str[1]
data['驱动电机转矩'] = data['驱动电机转矩'].str.split(':').str[1]
data['驱动电机温度'] = data['驱动电机温度'].str.split(':').str[1]
data['电机控制器直流母线电流'] = data['电机控制器直流母线电流'].str.split(':').str[1]
data['电机控制器输入电压'] = data['电机控制器输入电压'].str.split(':').str[1]
data.head()
```

```python
data.describe(include='all').T
```

```python
unique_values = set(data['充电状态'])
unique_values
```

```python
condition = data['充电状态'] == 1
data.loc[condition, ['驱动电机控制器温度', '驱动电机转速', '驱动电机转矩', '驱动电机温度', '电机控制器输入电压', '电机控制器直流母线电流']] = 0
```

```python
data['驱动电机控制器温度'] = pd.to_numeric(data['驱动电机控制器温度'], errors='coerce')
data['驱动电机转速'] = pd.to_numeric(data['驱动电机转速'], errors='coerce')
data['驱动电机转矩'] = pd.to_numeric(data['驱动电机转矩'], errors='coerce')
data['驱动电机温度'] = pd.to_numeric(data['驱动电机温度'], errors='coerce')
data['电机控制器输入电压'] = pd.to_numeric(data['电机控制器输入电压'], errors='coerce')
data['电机控制器直流母线电流'] = pd.to_numeric(data['电机控制器直流母线电流'], errors='coerce')
```

处理异常值

```python
data.describe(include='all').T
```

```python
unique_values = set(data['充电状态'])
unique_values
```

```python
data = data[data['累计里程'] <= 999999]
```

```python
data = data[data['驱动电机转速'] >= -1000]
```

```python
data = data[data['最高温度值'] >= 0]
data = data[data['最低温度值'] >= 0]
```

```python
data.describe(include='all').T
```

```python
data.to_csv('9个月总数据.csv', index = False)
```
